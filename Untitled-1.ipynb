```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
df = pd.read_csv("CreditPrediction.csv")
df.columns = df.columns.str.strip()


df = df.drop(['CLIENTNUM'], axis=1)# i deleted the first column because it was useless
df = df.drop(['Unnamed: 19'], axis=1)
df = df.drop(['Avg_Utilization_Ratio'], axis=1)

df.head()
```

```python
# simple feature engineering

df["Avg_Transaction_Amount"] = df["Total_Trans_Amt"] / (df["Total_Trans_Ct"] + 1)
df["Transactions_per_Month"] = df["Total_Trans_Ct"] / (df["Months_on_book"] + 1)
df["Amount_per_Month"] = df["Total_Trans_Amt"] / (df["Months_on_book"] + 1)
df["Inactive_Rate"] = df["Months_Inactive_12_mon"] / 12


numeric_columns = df.select_dtypes(include=['number']).columns
categorical_columns = df.select_dtypes(exclude=['number']).columns

# fill missing values
for col in numeric_columns:
    df[col] = df[col].fillna(df[col].median())

for col in categorical_columns:
    df[col] = df[col].fillna(df[col].mode()[0])

# ordinal encoding for ordered categorical columns
education_order = {
    "Unknown": 0,
    "Uneducated": 1,
    "High School": 2,
    "College": 3,
    "Graduate": 4,
    "Post-Graduate": 5,
    "Doctorate": 6
}

income_order = {
    "Unknown": 0,
    "Less than $40K": 1,
    "$40K - $60K": 2,
    "$60K - $80K": 3,
    "$80K - $120K": 4,
    "$120K +": 5
}

card_order = {
    "Blue": 1,
    "Silver": 2,
    "Gold": 3,
    "Platinum": 4
}

df["Education_Level"] = df["Education_Level"].map(education_order)
df["Income_Category"] = df["Income_Category"].map(income_order)
df["Card_Category"] = df["Card_Category"].map(card_order)

df["Education_Level"] = df["Education_Level"].fillna(0)
df["Income_Category"] = df["Income_Category"].fillna(0)
df["Card_Category"] = df["Card_Category"].fillna(1)


# one hot encoding for non-ordered columns
categorical_columns = df.select_dtypes(exclude=['number']).columns

df = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
```

در این مرحله، داده‌ها برای ورود به مدل‌های رگرسیونی آماده‌سازی شدند. از آنجا که الگوریتم‌های یادگیری ماشین با داده خام متنی یا مقدارهای خالی به خوبی کار نمی‌کنند، لازم بود قبل از آموزش مدل، مقادیر مفقود پر شوند و ویژگی‌های دسته‌ای به شکل عددی تبدیل شوند.

در ابتدا چند ویژگی جدید و ساده ساخته شد. برای مثال، ویژگی <b>Avg_Transaction_Amount</b> میانگین مبلغ هر تراکنش را نشان می‌دهد و از تقسیم مبلغ کل تراکنش‌ها بر تعداد تراکنش‌ها به دست آمده است. همچنین ویژگی‌هایی مثل <b>Transactions_per_Month</b> و <b>Amount_per_Month</b> ساخته شدند تا رفتار مشتری در طول مدت ارتباط با بانک بهتر مشخص شود. این ویژگی‌ها از ستون هدف یعنی <b>Credit_Limit</b> استفاده نمی‌کنند، بنابراین باعث نشت اطلاعات یا تقلب داده‌ای نمی‌شوند.

برای پر کردن مقادیر مفقود، ستون‌های عددی و دسته‌ای جداگانه بررسی شدند. در ستون‌های عددی از <b>میانه</b> استفاده شد، چون داده‌های مالی و بانکی معمولاً مقدارهای خیلی بزرگ یا خیلی کوچک دارند. اگر به جای میانه از میانگین استفاده شود، چند مقدار خیلی بالا یا پایین می‌توانند مقدار جایگزین را غیرواقعی کنند. اما میانه در برابر داده‌های پرت مقاوم‌تر است و برای این پروژه انتخاب مناسب‌تری محسوب می‌شود.

برای ستون‌های دسته‌ای مثل جنسیت، وضعیت تأهل یا نوع کارت، از <b>مد</b> یا همان پرتکرارترین مقدار استفاده شد. دلیل این انتخاب این است که این ستون‌ها عددی نیستند و مفاهیمی مثل میانگین یا میانه برای آن‌ها معنی ندارد. بنابراین منطقی‌ترین راه این است که مقدار خالی با رایج‌ترین مقدار همان ستون جایگزین شود.

بعد از مدیریت مقادیر مفقود، مرحله کدگذاری انجام شد. در این پروژه همه ویژگی‌های دسته‌ای به یک شکل کدگذاری نشدند، چون بعضی از آن‌ها دارای ترتیب طبیعی هستند و بعضی دیگر ترتیب خاصی ندارند.

برای ستون‌هایی مثل <b>Education_Level</b>، <b>Income_Category</b> و <b>Card_Category</b> از <b>Ordinal Encoding</b> استفاده شد. علت این انتخاب این است که این ویژگی‌ها دارای سطح و ترتیب هستند. برای مثال سطح تحصیلات از Uneducated تا Doctorate یک ترتیب منطقی دارد. همچنین سطح درآمد از Less than $40K تا $120K + به صورت طبیعی قابل مرتب‌سازی است. نوع کارت نیز از Blue تا Platinum نشان‌دهنده سطح متفاوتی از کارت بانکی است. اگر برای این ستون‌ها One-Hot Encoding استفاده می‌شد، مدل ترتیب بین سطح‌ها را متوجه نمی‌شد و همه حالت‌ها را کاملاً مستقل از هم در نظر می‌گرفت.

در مقابل، برای ستون‌هایی مثل <b>Gender</b> و <b>Marital_Status</b> از <b>One-Hot Encoding</b> استفاده شد. دلیل این کار این است که این ویژگی‌ها ترتیب واقعی ندارند. برای مثال نمی‌توان گفت Married از Single بزرگ‌تر یا کوچک‌تر است. بنابراین تبدیل آن‌ها به چند ستون صفر و یک روش مناسب‌تری است.

به طور کلی، هدف این مرحله این بود که داده‌ها بدون از دست رفتن معنی اصلی ویژگی‌ها، به شکل عددی و قابل استفاده برای مدل‌های رگرسیونی تبدیل شوند.

```python
#clean outliers without the y label which is credit limit
OUTLIER_METHOD = "IQR"


features_to_clean = [col for col in numeric_columns if col != 'Credit_Limit']

if OUTLIER_METHOD == "IQR":
    # 1. IQR Method
    for col in features_to_clean:
        if col in df.columns:
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1

            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

elif OUTLIER_METHOD == "Z-Score":
    # 2. Z-Score Method (using 3 standard deviations)
    for col in features_to_clean:
        if col in df.columns:
            mean = df[col].mean()
            std = df[col].std()

            lower_bound = mean - 3 * std
            upper_bound = mean + 3 * std

            df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

elif OUTLIER_METHOD == "IsolationForest":
    # 3. Isolation Forest Method
    from sklearn.ensemble import IsolationForest

    # contamination specifies the expected proportion of outliers (e.g., 5%)
    iso_forest = IsolationForest(contamination=0.05, random_state=42)
    outliers_preds = iso_forest.fit_predict(df[features_to_clean])

    # For Isolation Forest, instead of clipping, we clip the rows detected as outliers
    # by replacing extreme values with the feature's median to align with your pipeline
    for col in features_to_clean:
        median_val = df[col].median()
        # replace extreme values of detected outlier rows with median
        df.loc[outliers_preds == -1, col] = median_val

df = df.fillna(0)

print(f"Outliers processed successfully using standard {OUTLIER_METHOD} method.")
```

در این مرحله، داده‌های پرت یا همان Outliers در ستون‌های عددی بررسی و کنترل شدند. داده پرت به مقداری گفته می‌شود که نسبت به بیشتر داده‌های یک ستون فاصله زیادی دارد. در پروژه‌های مالی و بانکی، وجود داده پرت طبیعی است؛ چون ممکن است بعضی مشتریان تراکنش‌های بسیار بالا، تعداد تراکنش متفاوت یا رفتار اعتباری خاصی داشته باشند.

اگر داده‌های پرت بدون کنترل وارد مدل شوند، ممکن است مدل را به سمت الگوهای غیرواقعی ببرند. برای مثال اگر یک مشتری مقدار تراکنش بسیار بالایی داشته باشد، مدل ممکن است وزن زیادی به آن مقدار بدهد و در نتیجه برای مشتریان عادی خطای بیشتری تولید کند. به همین دلیل، کنترل داده‌های پرت باعث می‌شود مدل رفتار عمومی داده‌ها را بهتر یاد بگیرد.

در این پروژه از روش <b>IQR</b> استفاده شد. این روش بر اساس چارک اول و چارک سوم کار می‌کند. ابتدا بازه بین چارک اول و سوم محاسبه می‌شود و سپس حد پایین و حد بالا برای هر ویژگی به دست می‌آید. مقدارهایی که کمتر از حد پایین یا بیشتر از حد بالا باشند، به عنوان مقدارهای غیرعادی در نظر گرفته می‌شوند.

در این پروژه داده‌های پرت حذف نشدند، بلکه با روش <b>Clipping</b> محدود شدند. یعنی اگر یک مقدار از حد بالا بیشتر بود، همان مقدار به حد بالا تبدیل شد و اگر از حد پایین کمتر بود، به حد پایین تبدیل شد. این روش نسبت به حذف کردن ردیف‌ها مناسب‌تر است، چون باعث کم شدن حجم دیتاست نمی‌شود. از طرف دیگر، اثر مقدارهای خیلی غیرعادی را هم کنترل می‌کند.

نکته مهم این است که ستون <b>Credit_Limit</b> در این مرحله دستکاری نشد. چون این ستون متغیر هدف پروژه است و هدف مدل پیش‌بینی همین مقدار است. اگر مقدارهای هدف را تغییر دهیم، ممکن است مسئله اصلی پروژه تغییر کند و ارزیابی مدل واقعی نباشد. به همین دلیل فقط ویژگی‌های ورودی پاک‌سازی شدند و ستون هدف به همان شکل اصلی باقی ماند.

در نتیجه، این مرحله کمک می‌کند مدل‌ها روی داده‌هایی آموزش ببینند که هنوز اطلاعات اصلی خود را دارند، اما اثر مقدارهای بسیار غیرعادی در آن‌ها کمتر شده است.

```python
# extraction and scaling

X = df.drop('Credit_Limit', axis=1)
Y = df["Credit_Limit"]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# You can choose between: 'Standard' or 'MinMax'
SCALING_METHOD = "Standard"

if SCALING_METHOD == "Standard":    scaler = StandardScaler()
elif SCALING_METHOD == "MinMax":    scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

<h3 style="color: #4a90e2; border-bottom: 1px solid #333333; padding-bottom: 8px; margin-top: 0; font-size: 1.4em;">تحلیل پیش‌پردازش: جداسازی متغیر هدف و مقیاس‌گذاری ویژگی‌ها</h3>

در این مرحله، ستون <b>Credit_Limit</b> به عنوان متغیر هدف از سایر ستون‌ها جدا شد. در این پروژه هدف این است که مدل با استفاده از اطلاعات جمعیتی، رفتاری و بانکی مشتریان، مقدار سقف اعتبار کارت را پیش‌بینی کند. بنابراین ستون <b>Credit_Limit</b> به عنوان خروجی مدل و بقیه ستون‌ها به عنوان ورودی مدل در نظر گرفته شدند.

بعد از جداسازی متغیر هدف، داده‌ها به دو بخش آموزش و تست تقسیم شدند. ۸۰ درصد داده‌ها برای آموزش مدل و ۲۰ درصد برای ارزیابی مدل استفاده شد. این کار لازم است، چون اگر مدل فقط روی همان داده‌هایی ارزیابی شود که با آن‌ها آموزش دیده، نتیجه به دست آمده قابل اعتماد نیست. بخش تست کمک می‌کند عملکرد مدل روی داده‌هایی بررسی شود که قبلاً ندیده است.

در ادامه، مقیاس‌گذاری ویژگی‌ها انجام شد. دلیل این کار این است که ستون‌های مختلف دیتاست در مقیاس‌های متفاوتی قرار دارند. برای مثال، سن مشتری معمولاً عددی دو رقمی است، اما مبلغ تراکنش‌ها یا مانده حساب می‌تواند چند هزار واحد باشد. اگر این ویژگی‌ها بدون مقیاس‌گذاری وارد مدل شوند، بعضی الگوریتم‌ها ممکن است به ویژگی‌هایی با عددهای بزرگ‌تر اهمیت بیشتری بدهند، حتی اگر آن ویژگی واقعاً مهم‌تر نباشد.

در این پروژه از <b>StandardScaler</b> استفاده شد. این روش مقدارهای هر ویژگی را طوری تغییر می‌دهد که میانگین آن نزدیک صفر و انحراف معیار آن نزدیک یک شود. مزیت این روش این است که ویژگی‌ها را در یک مقیاس قابل مقایسه قرار می‌دهد و به خصوص برای مدل‌هایی مثل <b>Linear Regression</b> مناسب است.

البته مدل‌هایی مثل <b>Decision Tree</b> و <b>Random Forest</b> نسبت به مقیاس داده‌ها حساسیت کمتری دارند، چون بر اساس تقسیم‌بندی و شرط‌گذاری روی ویژگی‌ها کار می‌کنند. با این حال، چون در این پروژه چند مدل مختلف با هم مقایسه شدند، مقیاس‌گذاری روی همه داده‌های ورودی انجام شد تا شرایط آموزش برای مدل‌ها یکسان‌تر باشد.

نکته مهم دیگر این است که مقیاس‌گذاری بعد از تقسیم داده‌ها انجام شد. یعنی ابتدا داده‌ها به آموزش و تست تقسیم شدند و سپس <b>Scaler</b> فقط روی داده‌های آموزش fit شد. بعد همان تبدیل روی داده‌های تست اعمال شد. این کار از نشت اطلاعات بین داده تست و آموزش جلوگیری می‌کند و باعث می‌شود ارزیابی مدل واقعی‌تر باشد.

```python
#selection,training and validation for regression models


models = {
    "Linear Regression (the only Model that we currently know)": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(max_depth=6,min_samples_leaf=15, random_state=42),
    "Random Forest (Best model for this project, that certainly we dont know yet)": RandomForestRegressor(n_estimators=100,max_depth=10,min_samples_leaf=8, random_state=42)
}

predictions = {}

for name, model in models.items():
    model.fit(X_train_scaled, Y_train)
    Y_pred = model.predict(X_test_scaled)
    predictions[name] = Y_pred
    
    mse = mean_squared_error(Y_test, Y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(Y_test, Y_pred)
    
    print(f"--- {name} Performance ---")
    print(f"MSE:  {mse:.2f}")
    print(f"RMSE: {rmse:.2f}")
    print(f"R^2:  {r2:.4f}\n")
```

در این مرحله، پس از آماده‌سازی داده‌ها، چند مدل رگرسیونی برای پیش‌بینی مقدار <b>Credit_Limit</b> آموزش داده شدند. هدف از استفاده چند مدل این بود که فقط به یک روش وابسته نباشیم و بتوانیم ببینیم کدام مدل روی این دیتاست عملکرد بهتری دارد.

اولین مدل، <b>Linear Regression</b> بود. این مدل یکی از ساده‌ترین مدل‌های رگرسیونی است و فرض می‌کند رابطه بین ویژگی‌های ورودی و مقدار هدف تقریباً خطی است. استفاده از این مدل در این پروژه مفید است، چون یک مدل پایه به ما می‌دهد و می‌توانیم عملکرد مدل‌های دیگر را با آن مقایسه کنیم. اگر Linear Regression عملکرد ضعیف‌تری داشته باشد، می‌توان نتیجه گرفت که رابطه بین ویژگی‌ها و Credit_Limit کاملاً ساده و خطی نیست.

مدل دوم، <b>Decision Tree Regressor</b> بود. این مدل داده‌ها را به شاخه‌ها و شرط‌های مختلف تقسیم می‌کند. مزیت درخت تصمیم این است که می‌تواند رابطه‌های غیرخطی را بهتر از رگرسیون خطی تشخیص دهد. برای مثال ممکن است اثر سطح درآمد یا نوع کارت روی Credit_Limit در همه مشتریان یکسان نباشد. درخت تصمیم می‌تواند چنین تفاوت‌هایی را با تقسیم داده‌ها بهتر نشان دهد.

در این بخش، خطای باقی‌مانده یا <b>Residual</b> برای هر مدل بررسی شد. Residual اختلاف بین مقدار واقعی و مقدار پیش‌بینی‌شده است. یعنی اگر مقدار واقعی Credit_Limit با مقدار پیش‌بینی‌شده برابر باشد، خطا صفر است. هرچه مقدار residual از صفر دورتر شود، یعنی مدل در آن نمونه خطای بیشتری داشته است.

اگر residualها بیشتر اطراف خط صفر قرار بگیرند، یعنی مدل در بیشتر نمونه‌ها خطای زیادی ندارد. اما اگر خطاها در یک سمت بیشتر باشند، ممکن است مدل به صورت کلی مقدار Credit_Limit را کمتر یا بیشتر از واقعیت پیش‌بینی کند. همچنین اگر خطاها در بعضی قسمت‌ها خیلی بزرگ باشند، ممکن است نشان‌دهنده وجود نمونه‌های خاص یا الگوهایی باشد که مدل نتوانسته خوب یاد بگیرد.

در این پروژه، مدل Linear Regression خطاهای بیشتری نشان می‌دهد، که با مقدار RMSE بالاتر آن هماهنگ است. این یعنی مدل خطی برای همه نمونه‌ها توانایی کافی نداشته و احتمالاً رابطه بین ویژگی‌ها و Credit_Limit پیچیده‌تر از یک رابطه خطی ساده است.
