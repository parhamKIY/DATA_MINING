# Credit Limit Prediction Using Regression Algorithms

## Project Overview

The purpose of this project is to develop a machine learning regression model to predict customers' credit limits based on their demographic information, banking behavior, and transaction history.

The dataset contains different types of customer-related features such as age, income category, card type, transaction amount, transaction count, and account relationship information.

In this project, `Credit_Limit` is considered as the target variable, and different regression algorithms are implemented and compared to find a suitable prediction model.

The main steps of this project include:

- Data loading and initial exploration
- Data cleaning and preprocessing
- Handling missing values
- Feature engineering
- Outlier detection and treatment
- Feature analysis
- Model training
- Model evaluation and comparison

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

# Data Loading and Initial Data Cleaning

In this step, the dataset is loaded into a Pandas DataFrame and the initial quality of the data is examined.

Data cleaning is an important step in machine learning because incorrect, duplicated, or unnecessary data can negatively affect the performance of prediction models.

The following preprocessing operations are performed:

### Removing Empty Columns

Some columns may not contain useful information or may contain only missing values.
These columns are removed because they do not provide meaningful information for the prediction task.

### Removing Duplicate Records

Duplicate rows can cause the model to learn repeated patterns and may lead to biased evaluation results.

Therefore, duplicated samples are identified and removed from the dataset.

### Handling Duplicate Customers

The column `CLIENTNUM` is a unique identifier for each customer.
Duplicate customer IDs are checked to prevent having multiple records from the same customer.

### Removing Identifier Columns

Although `CLIENTNUM` identifies customers, it does not contain any predictive relationship with credit limit.
Therefore, it is removed before model training.

### Removing Data Leakage Features

The feature `Avg_Utilization_Ratio` is removed because it has a direct mathematical relationship with `Credit_Limit`.

Keeping this feature may allow the model to indirectly see information about the target variable, which is known as target leakage.

```python
df = pd.read_csv("CreditPrediction.csv")

df.drop(columns=["Unnamed: 19"], inplace=True)
#delete duplicate rows
df.drop_duplicates(inplace=True)
#if there are duplicate CLIENTNUM, keep the first one and drop the rest
df.drop_duplicates(subset="CLIENTNUM", keep="first", inplace=True)

df.drop(columns=["CLIENTNUM"], inplace=True)

df.drop(columns=["Avg_Utilization_Ratio"], inplace=True)

df.reset_index(drop=True, inplace=True)

df.head()
```

# Feature Engineering

Feature engineering is one of the important steps in machine learning projects because it helps to extract more meaningful information from the existing features.

In this step, new features are created based on customers' transaction behavior and banking activity.

The purpose of creating these features is to provide the machine learning models with additional information that may improve the prediction of the customer's credit limit.

The following features are generated:

### Average Transaction Amount

This feature represents the average amount of money spent in each transaction.

It is calculated by dividing the total transaction amount by the total number of transactions.

This feature helps the model understand the average spending behavior of each customer.

### Transactions Per Month

This feature represents the average number of transactions performed by a customer during their relationship period with the bank.

It provides information about customer activity level.

### Transaction Amount Per Month

This feature measures the average transaction volume of customers over the period of their relationship with the bank.

It can help identify customers with higher financial activity.

### Inactive Rate

This feature represents the ratio of inactive months during the last year.

It helps the model understand customer engagement and activity patterns.

Creating these additional features may reveal hidden patterns in customer behavior that are not directly available in the original dataset.

# Data Splitting and Missing Value Handling

Before training machine learning models, the dataset is divided into two parts:

- Training Data
- Testing Data

The training dataset is used to learn patterns from historical customer information, while the testing dataset is used to evaluate the model performance on unseen data.

The split is performed before applying preprocessing steps to prevent data leakage.

## Handling Missing Values

Real-world datasets often contain missing values due to incomplete information collection.

Missing values can negatively affect machine learning algorithms, therefore they must be handled before model training.

In this project:

- Numerical features are filled using the median value.
- Categorical features are filled using the most frequent value (mode).

Median is selected for numerical variables because it is less sensitive to extreme values compared to the mean.

The calculated values are obtained from the training dataset and then applied to the test dataset to maintain a fair evaluation process.

# Encoding Categorical Features

Machine learning algorithms require numerical inputs, therefore categorical variables must be converted into numerical representations.

In this project, categorical variables are handled using two approaches.

## Ordinal Encoding

Some categorical variables have a meaningful order between their categories.

For example:

- Education level
- Income category
- Card category

These variables are converted into numerical values according to their natural ranking.

This approach allows the model to understand the order relationship between categories.

## One-Hot Encoding

For categorical variables without a natural ranking, one-hot encoding is used.

This method creates separate binary columns for each category.

For example, instead of storing Gender as text values, separate numerical columns are created.

This prevents the model from assuming an incorrect mathematical relationship between unrelated categories.

After encoding, the training and testing datasets are aligned to ensure that both datasets have the same feature structure.

```python
# simple feature engineering

df["Avg_Transaction_Amount"] = (df["Total_Trans_Amt"] /df["Total_Trans_Ct"].replace(0, np.nan))
df["Transactions_per_Month"] = (df["Total_Trans_Ct"] /df["Months_on_book"].replace(0, np.nan))
df["Amount_per_Month"] = (df["Total_Trans_Amt"] /df["Months_on_book"].replace(0, np.nan))
df["Inactive_Rate"] = df["Months_Inactive_12_mon"] / 12


df.replace([np.inf, -np.inf], np.nan, inplace=True)

X = df.drop('Credit_Limit', axis=1)
Y = df["Credit_Limit"]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)


numeric_columns = X_train.select_dtypes(include=['number']).columns
categorical_columns = X_train.select_dtypes(exclude=['number']).columns

# fill missing values
for col in numeric_columns:
    median_val = X_train[col].median()
    X_train[col] = X_train[col].fillna(median_val)
    X_test[col] = X_test[col].fillna(median_val)

for col in categorical_columns:
    mode_val = X_train[col].mode()[0]
    X_train[col] = X_train[col].fillna(mode_val)
    X_test[col] = X_test[col].fillna(mode_val)

# ordinal encoding for ordered categorical columns
education_order = {
    "Unknown": 0,
    "Uneducated": 1,
    "High School": 2,
    "College": 3,
    "Graduate": 4,
    "Post-Graduate": 5,
    "Doctorate": 6
}

income_order = {
    "Unknown": 0,
    "Less than $40K": 1,
    "$40K - $60K": 2,
    "$60K - $80K": 3,
    "$80K - $120K": 4,
    "$120K +": 5
}

card_order = {
    "Blue": 1,
    "Silver": 2,
    "Gold": 3,
    "Platinum": 4
}

X_train["Education_Level"] = X_train["Education_Level"].map(education_order).fillna(0)
X_train["Income_Category"] = X_train["Income_Category"].map(income_order).fillna(0)
X_train["Card_Category"] = X_train["Card_Category"].map(card_order).fillna(1)

X_test["Education_Level"] = X_test["Education_Level"].map(education_order).fillna(0)
X_test["Income_Category"] = X_test["Income_Category"].map(income_order).fillna(0)
X_test["Card_Category"] = X_test["Card_Category"].map(card_order).fillna(1)


# one hot encoding for non-ordered columns
categorical_columns = X_train.select_dtypes(exclude=['number']).columns

X_train = pd.get_dummies(X_train, columns=categorical_columns, drop_first=True)
X_test = pd.get_dummies(X_test, columns=categorical_columns, drop_first=True)

X_train, X_test = X_train.align(X_test,join="left",axis=1,fill_value=0)
```

# Outlier Detection and Treatment

Outliers are observations that have values significantly different from the majority of the dataset.

In financial datasets, outliers may represent customers with unusual transaction behavior or extremely high credit limits.

Detecting and managing outliers is important because extreme values can affect the learning process of regression models.

In this project, three different methods are considered for detecting outliers:

## 1. Z-Score Method

The Z-score method measures how far a data point is from the average value of a feature.

It is calculated based on the mean and standard deviation of the data.

A large positive or negative Z-score indicates that the observation may be an outlier.

This method works better when the data approximately follows a normal distribution.

## 2. Interquartile Range (IQR) Method

The IQR method is based on quartiles and is commonly used for detecting outliers.

It calculates:

- Q1: The first quartile (25% of data)
- Q3: The third quartile (75% of data)
- IQR: Difference between Q3 and Q1

Observations outside the following range are considered potential outliers:

Lower Bound = Q1 - 1.5 × IQR

Upper Bound = Q3 + 1.5 × IQR

This method is more robust for datasets with skewed distributions.

## 3. Isolation Forest Method

Isolation Forest is a machine learning based anomaly detection algorithm.

It detects unusual observations by randomly separating data points.

Outliers are easier to isolate because they have different characteristics compared to normal observations.

After detecting extreme values, an appropriate treatment method is applied to reduce their negative effect on model performance.

```python
#clean outliers without the y label which is credit limit
OUTLIER_METHOD = "IQR"


features_to_clean = numeric_columns 

if OUTLIER_METHOD == "IQR":
    # 1. IQR Method
    for col in features_to_clean:
        if col in X_train.columns:
            Q1 = X_train[col].quantile(0.25)
            Q3 = X_train[col].quantile(0.75)
            IQR = Q3 - Q1

            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            X_train[col] = X_train[col].clip(lower=lower_bound, upper=upper_bound)
            X_test[col] = X_test[col].clip(lower=lower_bound, upper=upper_bound)

elif OUTLIER_METHOD == "Z-Score":
    # 2. Z-Score Method (using 3 standard deviations)
    for col in features_to_clean:
        if col in X_train.columns:
            mean = X_train[col].mean()
            std = X_train[col].std()

            lower_bound = mean - 3 * std
            upper_bound = mean + 3 * std

            X_train[col] = X_train[col].clip(lower=lower_bound, upper=upper_bound)
            X_test[col] = X_test[col].clip(lower=lower_bound, upper=upper_bound)

elif OUTLIER_METHOD == "IsolationForest":
    # 3. Isolation Forest Method
    from sklearn.ensemble import IsolationForest

    # contamination specifies the expected proportion of outliers (e.g., 5%)
    iso_forest = IsolationForest(contamination=0.05, random_state=42)
    outliers_preds_train = iso_forest.fit_predict(X_train[features_to_clean])
    outliers_preds_test = iso_forest.predict(X_test[features_to_clean])

    for col in features_to_clean:
        median_val = X_train[col].median()
        X_train.loc[outliers_preds_train == -1, col] = median_val
        X_test.loc[outliers_preds_test == -1, col] = median_val

X_train = X_train.fillna(0)
X_test = X_test.fillna(0)

print(f"Outliers processed successfully using standard {OUTLIER_METHOD} method.")
```

# Feature Correlation Analysis

Correlation analysis is performed to understand the relationship between numerical features.

The correlation coefficient shows the strength and direction of a linear relationship between two variables.

The value of correlation ranges from:

- +1 : Strong positive relationship
- 0 : No linear relationship
- -1 : Strong negative relationship

A correlation matrix is created and visualized using a heatmap.

The objectives of this step are:

- Understanding relationships between variables
- Identifying highly correlated features
- Detecting possible redundant information

Although highly correlated features are examined, they are not automatically removed because some machine learning algorithms, especially tree-based models, can handle correlated features effectively.

```python
# 1. Select independent features by dropping the target column 'Credit_Limit'
features_df = df.drop('Credit_Limit', axis=1)

# 2. Calculate the correlation matrix (using absolute values to capture both strong positive and negative correlations)
corr_matrix = X_train.corr().abs()

plt.figure(figsize=(14, 11))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Independent Features (Train Data)')
plt.tight_layout()
plt.show()

upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.85)]

print("Highly correlated features to be dropped:")
print(to_drop)

X_train = X_train.drop(columns=to_drop)
X_test = X_test.drop(columns=to_drop)
```

# Feature Scaling

Feature scaling is an important preprocessing step that helps machine learning algorithms work better when numerical features have different ranges.

In this dataset, features have different scales.

For example:

- Transaction amounts may have large numerical values.
- Transaction counts and customer-related features may have smaller values.

Without scaling, some algorithms that are based on distance or mathematical optimization may be affected by features with larger values.

In this project, StandardScaler is used for feature scaling.

StandardScaler transforms each feature by removing the mean and scaling it according to the standard deviation.

After scaling:

- The mean of each feature becomes approximately zero.
- The standard deviation becomes approximately one.

The scaler is fitted only on the training data and then applied to the test data.

This approach prevents information from the test dataset being used during the training process and helps provide a fair evaluation of model performance.

```python
# extraction and scaling


# You can choose between: 'Standard' or 'MinMax'
SCALING_METHOD = "Standard"

if SCALING_METHOD == "Standard":    scaler = StandardScaler()
elif SCALING_METHOD == "MinMax":    scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Regression Models Definition

In this step, different regression algorithms are defined to compare their ability to predict customers' credit limits.

Each algorithm has different learning characteristics and can capture different types of relationships between input features and the target variable.

The following regression models are used in this project:

## Linear Regression

Linear Regression is a basic statistical and machine learning algorithm that assumes a linear relationship between input variables and the target variable.

The goal of this algorithm is to find the best linear equation that minimizes the difference between actual and predicted values.

Advantages:

- Simple and easy to interpret
- Fast training process

Limitations:

- Cannot capture complex nonlinear relationships between features

---

## Decision Tree Regressor

Decision Tree is a supervised learning algorithm that predicts values by creating a tree-like structure based on decision rules.

The dataset is repeatedly divided into smaller groups according to feature conditions until the final prediction is obtained.

Advantages:

- Can model nonlinear relationships
- Easy to understand
- Does not require feature scaling

Limitations:

- May overfit when the tree becomes too complex

---

## Random Forest Regressor

Random Forest is an ensemble learning algorithm that combines multiple decision trees.

Instead of relying on a single tree, it creates several trees using different subsets of data and features.

The final prediction is obtained by averaging the predictions of all trees.

Advantages:

- Reduces overfitting compared to a single decision tree
- Handles complex relationships effectively
- Performs well on structured datasets

Limitations:

- Requires more computational resources
- Less interpretable than a single decision tree

---

## K-Nearest Neighbors (KNN) Regressor

K-Nearest Neighbors is a distance-based algorithm.

For a new observation, the model finds the closest training samples and predicts the output based on the values of those neighbors.

Advantages:

- Simple and easy to implement
- Does not require a complex training process

Limitations:

- Sensitive to feature scaling
- Prediction can become slower with larger datasets

These models are selected to provide a comparison between simple, tree-based, and distance-based regression approaches.

# Model Training and Performance Evaluation

In this step, all defined regression models are trained using the training dataset.

The training process allows each model to learn the relationship between customer characteristics and the target variable (`Credit_Limit`).

After training, each model predicts credit limits for the test dataset.

The performance of each model is evaluated using several regression metrics:

## Mean Absolute Error (MAE)

MAE measures the average absolute difference between actual values and predicted values.

A lower MAE indicates that the model predictions are closer to the real credit limits.

## Mean Squared Error (MSE)

MSE calculates the average squared difference between actual and predicted values.

Because errors are squared, larger prediction errors have a greater impact on this metric.

## Root Mean Squared Error (RMSE)

RMSE is the square root of MSE.

It has the same unit as the target variable, making it easier to interpret.

A lower RMSE represents better prediction accuracy.

## R² Score

R² score measures how much variation in the target variable can be explained by the model.

The closer the value is to 1, the better the model explains the changes in credit limit.

After calculating these metrics, the models are compared and the best-performing model is selected as the final prediction model.

# Regression Models Comparison

After training all regression models, the evaluation results are collected into a comparison table.

This table allows us to compare different algorithms based on:

- Prediction error (MAE and RMSE)
- Explained variance (R² Score)

The model with:

- Lower MAE
- Lower RMSE
- Higher R² Score

is considered the better model for predicting customer credit limits.

This comparison helps identify which algorithm is more suitable for this specific financial dataset.

# Final Model Selection and Prediction

Based on the evaluation results, the best-performing model is selected as the final prediction model.

The selected model is then used to generate predictions on the test dataset.

Using the final model, predicted credit limits are compared with the actual credit limits to evaluate how accurately the model performs on unseen data.

The final prediction results are used for further analysis, including visualization and feature importance evaluation.

```python
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "K-Nearest Neighbors": KNeighborsRegressor(n_neighbors=5)
}

results = {}
predictions = {}

for name, model in models.items():
    model.fit(X_train_scaled, Y_train)
    
    y_pred = model.predict(X_test_scaled)
    predictions[name] = y_pred
    mse = mean_squared_error(Y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(Y_test, y_pred)
    r2 = r2_score(Y_test, y_pred)
    
    results[name] = {
        "MSE": mse,
        "RMSE": rmse,
        "MAE": mae,
        "R2 Score": r2
    }
    
    print(f"--- {name} ---")
    print(f"R2 Score : {r2:.4f}")
    print(f"RMSE     : {rmse:.2f}")
    print(f"MAE      : {mae:.2f}\n")
```

# Feature Importance Analysis

Feature importance analysis is used to understand which input variables have the greatest influence on the model predictions.

In this project, Random Forest feature importance is used.

Random Forest calculates the contribution of each feature based on how much it improves the decision-making process across all trees.

Features with higher importance values have a stronger influence on predicting the customer's credit limit.

This analysis provides better understanding of customer characteristics and transaction behaviors that are most related to credit limit prediction.

```python
rf_model = models["Random Forest"]

importances = rf_model.feature_importances_
feature_names = X_train.columns
indices = np.argsort(importances)[::-1]

print("Top 5 important features:")
for i in range(5):
    print(f"{i+1}. {feature_names[indices[i]]}: {importances[indices[i]]:.4f}")

plt.figure(figsize=(9, 5))
plt.title("Top 5 Feature Importances (Random Forest)")
plt.bar(range(5), importances[indices][:5], align="center", color='teal', edgecolor='black')
plt.xticks(range(5), feature_names[indices][:5], rotation=15)
plt.tight_layout()
plt.show()
```

# Residual Analysis

Residuals represent the difference between actual values and predicted values.

They are calculated as:

Residual = Actual Value - Predicted Value

Residual analysis helps evaluate whether the model has systematic errors.

A good regression model should have residuals distributed randomly around zero.

Patterns in residuals may indicate problems such as:

- Missing important features
- Incorrect model assumptions
- Poor prediction performance for certain ranges of values

This analysis helps better understand the strengths and limitations of the final model.
